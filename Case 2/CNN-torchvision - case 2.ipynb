```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

num = 0
for root, dirs, files in os.walk("2nd dataset"):
    for file in files:
        print(os.path.join(root, file))
        num = num + 1
```

```python
import torch
from torch import nn
import torchvision 
import torchvision.transforms as transforms 
from torch.utils.tensorboard import SummaryWriter 
```

```python
writer = SummaryWriter()
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.densenet = torch.hub.load('pytorch/vision:v0.6.0', 'densenet201', pretrained=True)
        for p in self.densenet.parameters():
               p.requires_grad = False
        self.densenet.classifier = nn.Identity()
        self.flatten1 = nn.Flatten()
        self.linear1 = nn.Linear(1920,128)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.2)
        self.linear2 = nn.Linear(128,64)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(0.3)
        self.linear3 = nn.Linear(64,2)
        self.sigmoid = nn.Sigmoid()
    
    
    def forward(self,x):
        x=self.densenet(x)
        x=self.flatten1(x)
        x=self.linear1(x)
        x=self.relu1(x)
        x=self.dropout1(x)
        x=self.linear2(x)
        x=self.relu2(x)
        x=self.dropout2(x)
        x=self.linear3(x)
        x=self.sigmoid(x)
        return x
        
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
model = Net().to('cpu')
```

```python
model(torch.rand((1,3,224,224)))
```

```python
model.to(device)
```

```python
transform = transforms.Compose(
    [ transforms.Resize((224,224)),transforms.ToTensor(),
      transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5)),
    ])
```

```python
num
```

```python
dataset = torchvision.datasets.ImageFolder('2nd dataset', transform=transform)
train_size = int( num/10 * 8 )
val_size = int (num/10 ) 
test_size = int ( num - train_size - val_size )
train, val, test = torch.utils.data.random_split(dataset, [train_size, val_size, test_size])

trainloader = torch.utils.data.DataLoader(train,batch_size=50,shuffle=True,
                                           num_workers=1,drop_last = True)
valloader = torch.utils.data.DataLoader(val,batch_size=50,
                                           num_workers=1,drop_last = True)
testloader = torch.utils.data.DataLoader(test,batch_size=50,
                                           num_workers=1,drop_last = True)
images, labels = next(iter(trainloader))
```

```python
images
```

```python
import torch.optim as optim 

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(),lr=0.001, momentum=0.9)
```

```python
losses = []
test_accuracy_table = []
val_accuracy_table = []
min_valid_loss = np.inf

PATH = '2nd_trained_model.pt'


def accuracy(predicted, labels):
    classes = torch.argmax(predicted, dim=1)
    return torch.mean((classes == labels).float())*100

for epoch in range(35):
    
    train_loss = 0.0
    running_accuracy = 0.0
    val_accuracy = 0.0
    
    for i, data in enumerate(trainloader, 0):
        
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)
        
        
        optimizer.zero_grad()
       
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        train_loss += loss.item() * images.size(0) 
        running_accuracy += accuracy(outputs, labels)
        # print statistics
        # running_loss += loss.item()
        if i % 2000 == 1999:    # print every 2000 mini-batches
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, train_loss / 2000))
            train_loss = 0.0
        epoch_loss = train_loss / len(trainloader)
        losses.append(epoch_loss)
        print(f"Epoch:{epoch} Loss:{loss}")
    writer.add_scalar("Loss/train", epoch_loss, epoch)
    
    valid_loss = 0.0
    model.eval()     # Optional when not using Model Specific layer
    
    with torch.no_grad():
        
        for data, labels in valloader:
            if torch.cuda.is_available():
                data, labels = data.cuda(), labels.cuda()

            outputs = model(data)
            loss = criterion(outputs,labels)
            valid_loss = loss.item() * data.size(0)
            val_accuracy += accuracy(outputs, labels)

        running_accuracy /= len(trainloader)
        val_accuracy /= len(valloader)
        
        test_accuracy_table.append(running_accuracy)
        val_accuracy_table.append(val_accuracy)
        
        print(f'Epoch {epoch+1} \t Training Accuracy: {running_accuracy}% \t Validation Accuracy: {val_accuracy}%')
        if min_valid_loss > valid_loss:
            print(f'Validation Loss Decreased({min_valid_loss:.6f}--->{valid_loss:.6f}) \t Saving The Model')
            min_valid_loss = valid_loss
            # Saving State Dict
            torch.save(model.state_dict(), PATH)
        
print('Finished Training')

            
```

```python
the_model = Net()
the_model.load_state_dict(torch.load(PATH))
the_model = the_model.to(device)
```
